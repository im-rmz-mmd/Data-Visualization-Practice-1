```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1, 12, (2, 3))
x
```

```python
plt.plot(x)
plt.xlabel('Row')
plt.ylabel('Column')
```

```python
y = np.random.randint(1, 10, 5)
z = np.random.randint(11, 21, 5)
y, z
```

```python
plt.bar(z, y)
plt.xlabel('Column = y')
plt.ylabel('Row = z')
```

```python
df = pd.DataFrame({'Animals': ['cat', 'bee', 'dog', 'cat', 'wolf',
                               'cat', 'dog', 'sheep', 'fish', 'fish', 'cat',
                               'bee', 'dog', 'sheep']})
a = df.value_counts()

plt.figure(figsize=(10, 10))
plt.pie(a, labels= a.index, autopct= '%0.1f%%')
plt.xlabel('The Most Amount of Animals')
plt.legend()
```

```python
f = np.random.randint(25, 35, 5)
g = np.random.randint(45, 55, 5)

plt.scatter(f, g, color= 'green', marker= 'D')
plt.xlabel('Column = f')
plt.ylabel('Row = g')
f, g
```

```python
q = pd.DataFrame({'Age': [15, 20, 26],
                  'Salary': [310, 560, 102]})
w = pd.DataFrame({'Age': [15, 20, 26],
                  'Salary': [420, 250, 220]})

plt.bar(w['Age'], w['Salary'], color= 'green')
plt.bar(q['Age'], q['Salary'], color= 'red')
plt.xlabel('Salary Difference in Two Months')
```

```python
data = np.random.randint(1, 20, 5)
plt.pie(data, autopct= '%0.1f%%', labels= data, 
        explode= [0, 0, 0, 0, 0.1], shadow= True)
plt.xlabel('The Largest Number')
```
